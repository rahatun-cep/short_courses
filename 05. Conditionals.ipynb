# Conditionals

#### with the `if` keyword
In Julia, one way, two way and multi-way if statements are supported

```julia
if *condition 1*
    *option 1*
elseif *condition 2*
    *option 2*
else
    *option 3*
end
```

allows us to conditionally evaluate one of our options.
<br><br>
 

### One  way if conditional


````julia

println("Enter Temperature in integer or float form")

flush(stdout)
 
temp=readline()
temp=parse(Float64, temp)

if temp > 86
    println("It is hot!")
    println("Be sure to drink liquids.")
end
````

````julia
println("Enter age in integer or float form to determinie if you can vote")

flush(stdout)
 
age=readline()
age=parse(Float64, age)

if age >= 18
  
    println("Yes, you can vote ")
end
````

### two  way if conditional


````julia
println("Enter age in integer or float form to determinie if you can vote")

flush(stdout)
 
age=readline()
age=parse(Float64, age)

if age >= 18
  
    println("Yes, you can vote")
else 
     println("NO, you can not vote")
end
````

````julia
println("Enter Temperature in integer or float form")

flush(stdout)
 
temp=readline()
temp=parse(Float64, temp)

if temp > 86
    println("It is hot!")
    println("Be sure to drink liquids.")
else
    println("It is not hot.")
    println("Bring a jacket.")
end
println("Goodbye.")
````

# Example for Converting your raw score to letter grades

````julia
 
print("Enter raw score\n")
flush(stdout)

score = readline()
score = parse(Float64, score)

prefix="Your letter grade is\t"
if score < 60  
    println(prefix, "F")
    
elseif 60 <= score < 70
    println(prefix, "D")
    
elseif 70 <= score < 80
    println(prefix, "C")
    
elseif 80 <= score < 90
    println(prefix, "B")
else
    println(prefix, "A")
end

if score < 90
    println("\nGoodbye.")
else
    println("\nGreat Job!!!.")
end

````

### working with ternary operators

In Juli could also use the ternary operator with the syntax

```julia
a ? b : c
```

which equates to 

```julia
if a
    b
else
    c
end
```

Now let's say we want to return the larger of two numbers. Give `x` and `y` values here:

````julia
x = 54
y = 43
````

Using the `if` and `else` keywords, we might write:

````julia
if x > y
    x
else
    y
end
````

and as a ternary operator, the conditional looks like this:

````julia
(x > y) ? x : y
````

#### with short-circuit evaluation

We've already seen expressions with the syntax
```julia
a && b
```
to return true if both `a` and `b` are true. Of course, if `a` is false, Julia doesn't even need to know the value of `b` in order to determine that the overall result will be false. So Julia doesn't even need to check what `b` is; it can just "short-circuit" and immediately return `false`.  The second argument `b` might be a more complicated expression like a function call with a side-effect, in which case it won't even be called:

````julia
false && (println("hi"); true)
````

````julia
true && (println("hi"); true)
````

On the other hand, if `a` is true, Julia knows it can just return the value of `b` as the overall expression. This means that `b` doesn't necessarily need evaluate to `true` or `false`!  `b` could even be an error:

````julia
(x > 0) && error("x cannot be greater than 0")
````

Similarly, check out the `||` operator, which also uses short-circuit evaluation to perform the "or" operation.

````julia
true || println("hi")
````

and

````julia
false || println("hi")
````
